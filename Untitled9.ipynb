```python
# Import necessary libraries
import kagglehub

# Download latest version
path = kagglehub.dataset_download("himanshupoddar/zomato-bangalore-restaurants")

print("Path to dataset files:", path)
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from nltk.corpus import stopwords
import string
from sklearn.preprocessing import MinMaxScaler

# Loading the data
zomato_real = pd.read_csv("../input/zomato-bangalore-restaurants/zomato.csv")

# Preprocessing
zomato = zomato_real.drop(['url', 'dish_liked', 'phone'], axis=1)
zomato.drop_duplicates(inplace=True)
zomato.dropna(how='any', inplace=True)

# Rename columns for easier handling
zomato = zomato.rename(columns={'approx_cost(for two people)': 'cost', 'listed_in(type)': 'type', 'listed_in(city)': 'city'})

# Cleaning and converting data types
zomato['cost'] = zomato['cost'].astype(str).apply(lambda x: x.replace(',', '')).astype(float)
zomato = zomato.loc[zomato.rate != 'NEW']
zomato = zomato.loc[zomato.rate != '-'].reset_index(drop=True)
zomato['rate'] = zomato['rate'].apply(lambda x: x.replace('/5', '') if isinstance(x, str) else x).astype(float)

# Capitalize restaurant names for consistency
zomato.name = zomato.name.apply(lambda x: x.title())

# Encode 'Yes'/'No' to boolean values
zomato.online_order.replace(('Yes', 'No'), (True, False), inplace=True)
zomato.book_table.replace(('Yes', 'No'), (True, False), inplace=True)

# Calculate mean rating and normalize it
zomato['Mean Rating'] = zomato.groupby('name')['rate'].transform('mean')
scaler = MinMaxScaler(feature_range=(1, 5))
zomato[['Mean Rating']] = scaler.fit_transform(zomato[['Mean Rating']])

# Preprocess the reviews for similarity calculation
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

zomato['reviews_list'] = zomato['reviews_list'].str.lower()
zomato['reviews_list'] = zomato['reviews_list'].apply(remove_punctuation)

# Remove stopwords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

zomato["reviews_list"] = zomato["reviews_list"].apply(remove_stopwords)

# Function to filter by location, cuisine, and cost
def filter_restaurants(location, cuisine, max_cost):
    # Filter by location, cuisine, and cost range
    filtered = zomato[(zomato['city'] == location) &
                      (zomato['cuisines'].str.contains(cuisine, case=False, na=False)) &
                      (zomato['cost'] <= max_cost)]
    return filtered.reset_index(drop=True)

# Apply TF-IDF Vectorizer on reviews list for similarity calculations
tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0, stop_words='english')
filtered_restaurants_tfidf = None
cosine_similarities = None

def prepare_similarity_matrix(filtered):
    global filtered_restaurants_tfidf, cosine_similarities
    filtered_restaurants_tfidf = tfidf.fit_transform(filtered['reviews_list'])
    cosine_similarities = linear_kernel(filtered_restaurants_tfidf, filtered_restaurants_tfidf)
    indices = pd.Series(filtered.index, index=filtered['name']).drop_duplicates()
    return indices

# Recommendation function to find similar restaurants within filtered data
def recommend(indices, cosine_similarities):
    all_recommendations = []

    # Loop through all restaurants in the filtered list
    for idx in indices:
        recommend_restaurant = []
        score_series = pd.Series(cosine_similarities[idx]).sort_values(ascending=False)
        top_indexes = list(score_series.iloc[1:11].index)

        # Collect top recommendations based on review similarity and mean rating
        df_new = pd.DataFrame(columns=['name', 'cuisines', 'Mean Rating', 'cost'])
        for each in top_indexes:
            df_new = df_new.append(filtered.iloc[each][['name', 'cuisines', 'Mean Rating', 'cost']])

        # Append recommendations
        df_new = df_new.drop_duplicates().sort_values(by='Mean Rating', ascending=False).head(10)
        all_recommendations.append(df_new)

    return all_recommendations

# Main function to get recommendations
def get_recommendations(location, cuisine, max_cost):
    # Step 1: Filter restaurants by location, cuisine, and cost
    filtered = filter_restaurants(location, cuisine, max_cost)

    # Step 2: Prepare cosine similarity matrix for filtered data
    indices = prepare_similarity_matrix(filtered)

    # Step 3: Get recommendations
    recommended_restaurants = recommend(indices, cosine_similarities)
    print(f"Top recommended restaurants in '{location}' for '{cuisine}' cuisine under cost {max_cost}:")
    return recommended_restaurants

# Example usage
location = "Bangalore"
cuisine = "Italian"
max_cost = 1000
recommendations = get_recommendations(location, cuisine, max_cost)
print(recommendations)
```

```python


zomato_real = pd.read_csv(f"{path}/zomato.csv", usecols=['name', 'address', 'votes', 'rate', 'location', 'rest_type'])


zomato = zomato_real.drop_duplicates().dropna(subset=['name', 'address', 'votes', 'rate', 'location', 'rest_type'])


zomato = zomato[(zomato['rate'] != 'NEW') & (zomato['rate'] != '-')]

zomato['rate'] = zomato['rate'].apply(lambda x: x.replace('/5', '') if isinstance(x, str) else x).astype(float)

votes_data = np.array(zomato['votes']).reshape(-1, 1)
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(votes_data)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title("Elbow Method ")
plt.xlabel("Number of Clusters")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.show()

```

```python
optimal_clusters = 3

kmeans = KMeans(n_clusters=optimal_clusters, random_state=0)
zomato['cluster'] = kmeans.fit_predict(votes_data)

largest_cluster = zomato['cluster'].value_counts().idxmax()

zomato_largest_cluster = zomato[zomato['cluster'] == largest_cluster]

plt.figure(figsize=(10, 6))
sns.scatterplot(x=zomato['votes'], y=zomato['rate'], hue=zomato['cluster'], palette="viridis", legend="full")
plt.title("Clusters of Restaurants Based on Number of Votes")
plt.xlabel("Number of Votes")
plt.ylabel("Rating")
plt.show()

print("Restaurants in the Largest Cluster Based on Votes:")
display(zomato_largest_cluster[['name', 'address', 'votes', 'rate', 'location', 'rest_type']])
```
